```python
# South and west non profits
s_df = pd.read_csv('https://www.irs.gov/pub/irs-soi/eo3.csv')
s_df.head()
```

```python
all_df = pd.concat([s_df, e_df, ne_df])

all_names = list(all_df.groupby('STATE')['NAME'].nunique())

all_states = list(all_df.groupby('STATE').nunique().index)
```

```python
import plotly.graph_objects as go

def plot_us(states, items, colortitle, imagetitle):
  fig = go.Figure(data=go.Choropleth(
      locations=states,
      z = items,
      locationmode = 'USA-states', # set of locations match entries in `locations`
      #colorscale = 'Reds',
      colorbar_title = colortitle
  ))

  fig.update_layout(
      title_text = imagetitle,
      geo_scope='usa', # limite map scope to USA
  )

  return fig.show()

plot_us(all_states, all_names, "Tax-exempt entities", 'Tax-exempt entities by state' )
```

```python
all_df.groupby('STATE')['ACTIVITY'].head()
```

```python
all_income = list(all_df.groupby('STATE')['INCOME_AMT'].nunique())
all_income
```

```python
plot_us(all_states, all_income, "Tax-exempt income", 'Tax-exempt entities that actually have income by state' )
```
